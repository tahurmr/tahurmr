```python
# Ordinary Least Square Method
```

```python
#step1 = import library
import pandas as pd
```

```python
#step2 = import data
salary = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
salary.head()
```

```python
salary.info()
```

```python
salary.describe()
```

```python
#step3 = define target(y) and features(x)
```

```python
salary.columns
```

```python
y = salary['Salary']
x = salary[['Experience Years']]
```

```python
#step4 = split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=2529)  #random_state used for constantly showing same result
```

```python
# step5 = select model
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#step6 = train model (fit model)
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# so now we can find/predict salary... salary = 27010(intercept) + 9343(coefficient)*Experience {y = c + mx}
#so based on experience we can predict salary..
#suppose a person has 1 year experience.. so it becomes.
# salary = 27010 + 9343 * 1 = 33243
```

```python
# step7 = model prediction
y_pred = model.predict(x_test)
```

```python
x_test
```

```python
# step8 = accuracy
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
